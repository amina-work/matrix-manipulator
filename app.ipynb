```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

var = pd.read_csv("matrix.csv")

fig, ax = plt.subplots(5, 2)
ax[0, 0].plot(var["Position"], var["CTCF"], 'tab:blue')
ax[0, 0].set_title("CTCF", loc='left')
ax[0, 1].plot(var["Position"], var["H3K4me1"], 'tab:blue')
ax[0, 1].set_title("H3K4me1", loc='left')
ax[1, 0].plot(var["Position"], var["EZH2"], 'tab:green')
ax[1, 0].set_title("EZH2", loc='left')
ax[1, 1].plot(var["Position"], var["ATAC"], 'tab:green')
ax[1, 1].set_title("ATAC", loc='left')
ax[2, 0].plot(var["Position"], var["POL2"], 'tab:red')
ax[2, 0].set_title("POL2", loc='left')
ax[2, 1].plot(var["Position"], var["P300"], 'tab:red')
ax[2, 1].set_title("P300", loc='left')
ax[3, 0].plot(var["Position"], var["H3K9me3"], 'tab:cyan')
ax[3, 0].set_title("H3K9me3", loc='left')
ax[3, 1].plot(var["Position"], var["H3K4me3"], 'tab:cyan')
ax[3, 1].set_title("H3K4me3", loc='left')
ax[4, 0].plot(var["Position"], var["H3K36me3"], 'tab:purple')
ax[4, 0].set_title("H3K36me3", loc='left')
ax[4, 1].plot(var["Position"], var["H3K27ac"], 'tab:purple')
ax[4, 1].set_title("H3K27ac", loc='left')
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=2)
plt.show()
```
